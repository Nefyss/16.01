```python
import pandas as pd
import numpy as np
```

```python
ptd1 = pd.read_csv('part_01.csv')
ptd2 = pd.read_csv('part_02.csv')
ptd3 = pd.read_csv('part_03.csv')
ptd4 = pd.read_csv('part_04.csv')
ptd5 = pd.read_csv('part_05.csv')
ptd6 = pd.read_csv('part_06.csv')
ptd7 = pd.read_csv('part_07.csv')
ptd8 = pd.read_csv('part_08.csv')
ptd9 = pd.read_csv('part_09.csv')
ptd10 = pd.read_csv('part_10.csv')
ptd11 = pd.read_csv('part_11.csv')
ptd12 = pd.read_csv('part_12.csv')

ptd1['ID'].unique
```

```python
m1 = pd.concat([ptd5, ptd12], ignore_index=True)
m2 = pd.concat([ptd6, ptd10], ignore_index=True)
m3 = pd.concat([ptd2, ptd11], ignore_index=True)
m4 = pd.concat([ptd1, ptd9], ignore_index=True)
m5 = pd.concat([ptd3, ptd8], ignore_index=True)
m6 = pd.concat([ptd4, ptd7], ignore_index=True)
```

```python
df1 = m1.merge(m2, left_on='ID', right_on='ID')
df1 = df1.merge(m3, left_on='ID', right_on='ID')
df1 = df1.merge(m4, left_on='ID', right_on='ID')
df1 = df1.merge(m5, left_on='ID', right_on='ID')
df1 = df1.merge(m6, left_on='ID', right_on='ID')
df1 = df1.sort_values(by='ID')
df1
```
